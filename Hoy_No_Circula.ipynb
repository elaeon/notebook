# El programa Hoy no Circula aplicado de forma permanente no es mejor que uno Temporal.

##Motivos
Este texto trata de mostrar que el programa Hoy no Circula(HNC) en todas sus variantes no aportaría nada más a lo que haría un programa Hoy no Circula aplicado de forma temporal.

##Introducción
En la Ciudad de México la implantación del programa Hoy no Circula inicia el 20 de noviembre de 1989 de forma temporal(solo aplicada cuando se tiene contingencia ambiental o en el invierno), esto con el fin de reducir en un 20% la circulación de autos particulares, y así menguar la contaminación medida en el aire. Este programa duró 1 año de forma temporal, y a finales de 1990 se implemento de forma permanente. 

Desde entonces el Hoy no Circula ha sufrido diversos cambios [http://www.eluniversal.com.mx/notas/518895.html] entre los más relevantes están:

El 01 de septiembre del 2008 dentro del programa Hoy no Circula se implementó el Hoy no Circula sabatino, agregando las siguientes restricciones: 
 * Un sábado de cada mes dejan de circular los vehículos con holograma "2" o sin verificación vehicular (autos de otros estados), 
 * Dismunución de los años del automovil para tener el holograma cero (de 10 años a 8 años).

A partir de julio de 2014 se realizo otra modificación:
 * Dos sábados de cada mes dejan de circular los vehículos con holograma "1" (antes holograma "2").
 * Deján de circular todos los sábados de cada mes los vehículos con holograma "2" o sin verificación vehicular.

##Resultados entregados del Hoy No Circula
Como lo plantea este análisis [http://www.fisica.unam.mx/personales/hgriveros/docu/10-HoyNoCircula.299122533.pdf] el programa Hoy no Circula no entrega los resultados esperados, solo mejora de manera temporal la circulación, que es lo mismo que haría si funcionara temporalmente el programa.

La siguiente es una gráfica que nos muestra como cada año el número de automoviles particulares registrados en circulación va comportandose:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import collections
FIGSIZE_NORMAL = (15,6)
autos_arc = pd.read_csv('ARC.csv')
autos_arc_t = autos_arc.sum()
autos_arc_t['1980'] = 1601867
autos_arc_t['1981'] = 1706435
autos_arc_t['1982'] = 1707384
autos_arc_t['1983'] = 1542868
autos_arc_t['1984'] = 1564909
autos_arc_t['1985'] = 1614986
autos_arc_t['1986'] = 1359962
autos_arc_t['1987'] = 1388280
autos_arc_t['1988'] = 1508239
autos_arc_t['1989'] = 1538193
autos_arc_t['1990'] = 1768683
autos_arc_t['1991'] = 1727836
autos_arc_t['1992'] = 2019494
autos_arc_t['1993'] = 2314499
autos_arc_t = autos_arc_t.reindex_axis(sorted(autos_arc_t.keys()), axis=0)
autos_arc_t.plot(figsize=FIGSIZE_NORMAL, style='bs-').annotate('HNC Temporal',
             xy=(9, autos_arc_t['1989']),
             xytext=(40, 0), 
             textcoords='offset points',
             arrowprops=dict(facecolor='black', shrink=0.05))
autos_arc_t.plot(figsize=FIGSIZE_NORMAL, style='bs-').annotate('HNC Permanente',
             xy=(10, autos_arc_t['1990']),
             xytext=(50, 80), 
             textcoords='offset points',
             arrowprops=dict(facecolor='black', shrink=0.05))
autos_arc_t.plot(figsize=FIGSIZE_NORMAL, style='bs-').annotate('HNC Sabatino',
             xy=(28, autos_arc_t['2008']),
             xytext=(50, 20), 
             textcoords='offset points',
             arrowprops=dict(facecolor='black', shrink=0.05))
autos_arc_t.plot(figsize=FIGSIZE_NORMAL, style='bs-').annotate('Crisis de 1994',
             xy=(14, autos_arc_t['1994']),
             xytext=(50, 80), 
             textcoords='offset points',
             arrowprops=dict(facecolor='black', shrink=0.05))
```

Como podemos apreciar en la gráfica, de 1990 a 1993 hay un comportamiento completamente diferente al que se ve de 1980 a 1989.

Y bueno la cifra alarmante se empieza a notar de 2006 a 2012.

Una pregunta que nos viene a la mente es ¿a que se debe ese incremento tan abrupto desde el año 2006? si la población del 2005 al 2010 tuvo una tasa de crecimiento de 0.74% (130,164 personas) en cinco años, mientrás que el número de automobiles tuvo una tasa de crecimeinto de 21.78% (1,090,973 autos). 

Primero entendamos que cuando uno adquiere un auto, este tiene un periodo de vida, en ese periodo de vida es cuando se da el aumento del parque automotriz (al terminar ese periodo se da la dismunición del parque automotriz).

De acuerdo al número de autos nuevos vendidos al público que nos ofrece el INEGI (que data de 1994 a 2010), podemos obtener una gráfica que nos muestra la liberación del parque automotriz, junto con los autos que no fueron comprados dentro del D.F, pero están registrados en el D.F (foráneos).

```python
autos_n = pd.read_csv('ANC.csv')
autos_sin_nuevos = autos_arc_t - autos_n
order_dict = collections.OrderedDict()
def eliminar_acumulado(year, year_base, autos_sin_nuevos, autos_n):
    acum = 0
    for i in range(year - year_base):
        acum += autos_n[str(year - i - 1)]
    return autos_sin_nuevos[str(year)] - acum

for year in range(1994, 2011):
    order_dict[str(year)] = eliminar_acumulado(year, 1994, autos_sin_nuevos, autos_n)

autos_sin_nuevos1994_2010 = pd.DataFrame(order_dict)
autos_sin_nuevos1994_2010.T.diff().plot(figsize=FIGSIZE_NORMAL, kind='bar', legend=False)
```

Ahora comparemos esta gráfica con la anterior.

Las cifras negativas son años en que la adquisición de autos foráneos no excedió el desecho de autos (término de vida o ya no está registrado en el D.F). Las cifras positivas son años en que hubo más adquisición de autos foráneos y menor desecho de autos. 

En el 2006 se desecharon menos autos en comparación a la adquisición, esto también ocurrio en 2007, 2009, 2010 y deduciendo de la gráfica anterior, 2011 y 2012 tuvieron el mismo comportamiento. Esto nos lleva a inferir que la población con automóvil no está deshaciendose de su auto viejo, sino que además adquieren más autos.

#Conclusiones

Desde una perspectiva analítica vemos que el programa Hoy No Circula, cualquiera que sea su variante, no estimula el retiro de autos viejos de la circulación, estimula la adquisición de autos nuevos, seminuevos y foráneos. Una población que crece 0.74% contra el 21.78% de los autos en cinco años, nos indica que las familias con auto (46.4% de las familias con hogar tiene auto) adquieren otro auto, mientras que son pocas las familias que adquieren un auto por primera vez. 

El objetivo del HNC es retirar el 20% de los autos para el mejoramiento de la calidad del aire, sin embargo no está siendo cumplida (se calcula que es de un 8% a 10%), además que estos autos retirados son de las familias con menos posibilidad de adquirir uno nuevo.

¿Cómo se puede llegar al nivel mínimo de congestionamiento vial?

El congestinamiento vial puede ser disminuido al mínimo si se tiene alternativas de transporte que utilicen su propia via.

Agrandes rasgos, el transporte principal debe de ser el masivo y las bicicletas
  * El masivo en zonas de alta demanda deben ser trenes.
  * El masivo en zonas de media-alta demanda deben ser camiones articulados.
  * El masivo en zonas de baja-media demanda deben ser camiones con capacidad menor a 60 personas

El transporte secundario deben ser taxis y automoviles particulares.

Cuando las personas tengan en su vida cotidiana estas prioridades de transporte (con base en acciones de mejoramiento del transporte por parte del gobierno) se llegará a una mejor compartición de los recursos viales.

```python
#Herramientas para obtener la tasa de crecimiento
habitantes = pd.read_csv('HDF.csv')
def tasa_crecimiento(df, year_i, year_p, columns, n):
    tasa_c = collections.OrderedDict()
    for col in columns:
        tasa_c[col] = (((df[year_i][col] / float(df[year_p][col]))**(1.0/n)) - 1) * 100
    return pd.DataFrame(tasa_c, index=[0])

def tasa_crecimiento_t(df, year_i, year_p, n):
    tasa_c = (((df[year_i].sum() / float(df[year_p].sum()))**(1.0/n)) - 1) * 100
    return pd.DataFrame({0: tasa_c}, index=[0])

print tasa_crecimiento_t(autos_arc_t, '2010', '2005', 2)
habitantes_T = habitantes.T
habitantes_T.columns = columns
print tasa_crecimiento_t(habitantes_T.T, '2010', '2005', 2)
```
